```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
single_muts = pd.read_csv('single_muts_train.csv')
#print(single.head())
data_s = {}

for index, row in single_muts.iterrows():
  topology = str(row['name']).split('_',1)[0]
  if topology in data_s.keys():
    data_s[topology] += 1
  else:
    data_s[topology] = 1 

print("Topology-wise distribution:",data_s)
print("Total:", sum(data_s.values()))
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
ax.pie(list(data_s.values()), labels = list(data_s.keys()), autopct='%1.2f%%')
  
plt.title(label="Percentage of different topology for single mutation data",
          loc="center",
          fontstyle='italic')
plt.show()
```

```python
def addlabels(a,b):
    for i in range(len(a)):
        plt.text(i,b[i], b[i])

keys, values = data_s.keys(), data_s.values()

plt.title(label='Single mutation data')
plt.bar(keys, values)
plt.xlabel("Topology")
plt.ylabel("Number of entries")

addlabels(list(keys), list(values))
plt.show()
```

```python
multi_muts = pd.read_csv('multiple_muts_train.csv')

data_m = {}
for index, row in multi_muts.iterrows():
  topology = str(row['name']).split('_',1)[0]
  if topology in data_m.keys():
    data_m[topology] += 1
  else:
    data_m[topology] = 1 

print("Topology-wise distribution:",data_m)
print("Total:", sum(data_m.values()))

for one in data_m:
  print(one,":",data_m[one]*100/sum(data_m.values()))
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
ax.pie(list(data_m.values()), labels = list(data_m.keys()), autopct='%1.2f%%')
  
plt.title(label="Percentage of different topology for multi mutation data",
          loc="center",
          fontstyle='italic')
plt.show()
```

```python
keys, values = data_m.keys(), data_m.values()
plt.title(label='multi mutation data')
plt.bar(keys, values)
addlabels(list(keys), list(values))
plt.xlabel("Topology")
plt.ylabel("Number of entries")
plt.show()
```
